```
#Connect Colab to Google Drive
from google.colab import drive

#Mount Google Drive to access files
drive.mount("/content/drive")
```

```
# specify input folder
input_folder = "/content/drive/MyDrive/Ersilia"

# specify output folder
output_folder = "/content/drive/MyDrive/Ersilia"

# specify the name of the sample dataset to be used to test the model
input_filename = "test_model_eos2ta5"

# specify the SMILES column name
smiles_column = "smiles"
```

```
# Extract SMILES to a list
import pandas as pd

path = "%s/%s.xlsx" % (input_folder, input_filename)
smi_col = "%s" % smiles_column
df = pd.read_excel(path)
smiles = df[smi_col].tolist()

print("My dataset contains" + " " + str(len(smiles)) + " " + "SMILES")
```

```
# specify the model name
model_name = "eos2ta5"
```

```
# Fetch the model
import time

begin = time.time()
!ersilia fetch $model_name
end = time.time()

print("Time taken:", round((end - begin), 2), "seconds")
```

```
# serving the model
!ersilia serve $model_name
```

```
# Run predictions
api = "calculate"

from ersilia import ErsiliaModel
import time

model = ErsiliaModel(model_name)
begin = time.time()
output = model.api(input=smiles, output="pandas")
end = time.time()

print("Successful üëç! Time taken:", round((end - begin), 2), "seconds")
model.close()
```

```
# Check your results

print(output.head())
```

```
#Saving the result to Google Drive

output_path = "%s/%s_output.csv" % (output_folder, model_name)
output.to_csv(output_path, index=False)
```
